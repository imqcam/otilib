```python
#*****************************************************************************************************
def solve_2d_problem_prealloc(Th):
    
    ndim_analysis = 2
    els = Th.elements[2]

    fem.end_elements()
    
    K = oti.lil_matso((Th.x.shape[0],Th.x.shape[0]))
    f = oti.zeros((Th.x.shape[0],1))
    funct = 2*(Th.x**2+Th.y**2-Th.x-Th.y)
    # 
    for j in range(els['types'].size):

        elem = fem.element[ els['types'][j] ]

        if not elem.is_allocated():

            elem.end()
            elem.allocate(intorder=2)
            elem.allocate_spatial(ndim_analysis,compute_Jinv = True)

            fh = oti.zeros( ( elem.nbasis, 1 ) )
            NN_tmp1 = oti.zeros( ( elem.nbasis, elem.nbasis ) )
            NN_tmp2 = oti.zeros( ( elem.nbasis, elem.nbasis ) )
            NT_tmp1 = oti.zeros( ( elem.nbasis, 1 ) )
            NT_tmp2 = oti.zeros( ( elem.nbasis, 1 ) )
        # end if 

        elm_nodes = els['indices'][j]

        for i in range(elm_nodes.shape[0]):
            
            # reinitialize values.
            NN_tmp1.set(0.0)
            NN_tmp2.set(0.0)
            NT_tmp1.set(0.0)
            elem.set_coordinates(Th.x,Th.y,Th.z,elm_nodes[i,:])
            elem.compute_jacobian()

            Nx = elem.Nx
            Ny = elem.Ny
            N  = elem.N
            
            for ii in range(elem.nbasis):
                fh[ii,0] = funct[ int(elm_nodes[i,ii]),0]
            # end for 
            
            oti.transpose(Nx,out=NT_tmp1)
            oti.transpose(Ny,out=NT_tmp2)
            
            oti.dot(oti.transpose(Nx),Nx)
            
            tmp = - oti.dot(oti.transpose(Nx),Nx) - oti.dot(oti.transpose(Ny),Ny) 
            
            Kel = oti.gauss_integrate(tmp,elem.w_dJ)
            
            tmp = oti.dot_product(fh,N)
            fel = oti.gauss_integrate(tmp*oti.transpose(N),elem.w_dJ)
          
            assemble_globals(Kel,fel,K,f,elm_nodes[i,:])
 
        # end for

    # end for

    fem.end_elements()   
    
    # Setting Dirichlet BCs using TGV.
    TGV = 1e30
    
    # Every 1D node has dirichlet bc = 0
    els = Th.elements[1]
    for j in range(els['types'].size):

        elm_nodes = np.unique(els['indices'][j])

        for ii_ in elm_nodes:
            ii = int(ii_)
            K[ii,ii] = TGV
            # end for
            f[ii,0] = 0.0
        # end for 
        
    # end for
    
#     print(K.real.todense())
    u = oti.solve(K.tocsr(),f)

    return u,K,f

#-----------------------------------------------------------------------------------------------------
```

```python
#*****************************************************************************************************
def solve_2d_problem(Th):
    
    ndim_analysis = 2
    els = Th.elements[2]

    fem.end_elements()
    
    K = oti.lil_matso((Th.x.shape[0],Th.x.shape[0]))
    f = oti.zeros((Th.x.shape[0],1))
    funct = 2*(Th.x**2+Th.y**2-Th.x-Th.y)
    # 
    for j in range(els['types'].size):

        elem = fem.element[ els['types'][j] ]

        if not elem.is_allocated():

            elem.end()
            elem.allocate(intorder=2)
            elem.allocate_spatial(ndim_analysis,compute_Jinv = True)

            fh = oti.zeros( ( elem.nbasis, 1 ) )
            
        # end if 

        elm_nodes = els['indices'][j]

        for i in range(elm_nodes.shape[0]):
            
            # reinitialize values.

            elem.set_coordinates(Th.x,Th.y,Th.z,elm_nodes[i,:])
            elem.compute_jacobian()

            Nx = elem.Nx
            Ny = elem.Ny
            N  = elem.N
            
            for ii in range(elem.nbasis):
                fh[ii,0] = funct[ int(elm_nodes[i,ii]),0]
            # end for 
            
            tmp = - oti.dot(oti.transpose(Nx),Nx) - oti.dot(oti.transpose(Ny),Ny) 
            
            Kel = oti.gauss_integrate(tmp,elem.w_dJ)
            
            tmp = oti.dot_product(fh,N)
            fel = oti.gauss_integrate(tmp*oti.transpose(N),elem.w_dJ)
          
            assemble_globals(Kel,fel,K,f,elm_nodes[i,:])
 
        # end for

    # end for

    fem.end_elements()   
    
    # Setting Dirichlet BCs using TGV.
    TGV = 1e30
    
    # Every 1D node has dirichlet bc = 0
    els = Th.elements[1]
    for j in range(els['types'].size):

        elm_nodes = np.unique(els['indices'][j])

        for ii_ in elm_nodes:
            ii = int(ii_)
            K[ii,ii] = TGV
            # end for
            f[ii,0] = 0.0
        # end for 
        
    # end for
    
#     print(K.real.todense())
    u = oti.solve(K.tocsr(),f)

    return u,K,f

#-----------------------------------------------------------------------------------------------------
    
def assemble_globals(Kel,fel,Kgl,fgl,elem):
    
    
    for i in range(elem.size):
        
        ii=int(elem[i])
        
        for j in range(elem.size):

            jj=int(elem[j])
        
            Kgl[ii,jj] = Kgl[ii,jj] + Kel[i,j]
            
        # end for 
        fgl[ii,0] = fgl[ii,0] + fel[i,0]
        
    # end for 

    
def analytic_solution(Th):
    return Th.x*(1-Th.x)*Th.y*(1-Th.y)
```

```python
u,K,f = solve_2d_problem(Th)
# print(u)

u_analytic = analytic_solution(Th)
```
