```python
def rk45( f, t_start, t_end, y_init, h_init, tol = 1e-5, fargs = (),verb = 0 ):

    epsilon = mdl.mdual(tol)
    h       = h_init
    t       = t_start#+0*mdl.eps(1)
    w       = y_init
    i       = 0
    
    if verb > 2:
        print('Step {0:<10d}: t = {1:<15.8f}, w = {2:<15.8f}'.format(i,t,w) )
    # end if 

    Y = [w]
    T = [t]

    while (t<t_end):
    
        h = min(h, t_end-t)
        
        k1 = h*f(t,w,*fargs)
        k2 = h*f(t+h/4.0, w+k1/4.0,*fargs)
        k3 = h*f(t+ 3.0*h/ 8.0, w+   3.0*k1/  32.0+   9.0*k2/  32.0,*fargs)
        k4 = h*f(t+12.0*h/13.0, w+1932.0*k1/2197.0-7200.0*k2/2197.0+7296.0*k3/2197.0,*fargs)
        k5 = h*f(t+h, w+439*k1/216-8*k2+3680*k3/513-845*k4/4104,*fargs)
        k6 = h*f(t+h/2, w-8*k1/27+2*k2-3544*k3/2565+1859*k4/4104-11*k5/40,*fargs)
        
        w1 = w + 25.0*k1/216.0+1408.0*k3/ 2565.0+ 2197.0*k4/ 4104.0-k5/5.0
        w2 = w + 16.0*k1/135.0+6656.0*k3/12825.0+28561.0*k4/56430.0-9.0*k5/50.0+2.0*k6/55.0
        
        if verb >1:
            print("k1: ", k1)
            print("k2: ", k2)
            print("k3: ", k3)
            print("k4: ", k4)
            print("k5: ", k5)
            print("k6: ", k6)
            print("w1: ",w1 )
            print("w2: ",w2 )
            print("t : ",t )
            print()
        # end if 
        
        R = rk.mdl_abs(w1-w2)/h;
        delta = 0.84*(epsilon/R)**(0.25);
        # print(delta)
        if np.isnan(w2[0]):
            print('error. Got NaN as an answer.')
            break
        if rk.mdl_leq(R,epsilon):
        
            t = t+h
            w = w1
            i = i+1
            
            if verb > 2:
                print('Step {0:<10d}: t = {1:<15.8f}, w = {2:<15.8f}'.format(i,t,w) )
            # end if 
        
            Y.append(w)
            T.append(t)
            h = (delta*h)[0]
        
        else:
        
            h = (delta*h)[0]
        
        # end if


    # end while

    return (T,Y)
# end function
```
