```python
# How to determine the number of multiplication tables given the order:

for n in range(1,11):
    print()
    n_multtabl = n//2
    print("For order",n,", the total number of multtabl is", n_multtabl, "and are subdivided as follows:")
    for i in range(1,n//2+1):
        print("->",i,"+", n-i, "is", n)
```
