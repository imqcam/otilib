```python
import sympy as sym
from sympy.abc import i,j,k
```

```python
i*j
```

```python
expr = (k-(j-1)-(i-1)*(i-1+1)/2).expand()
expr
```

```python
sol= sym.solve(expr,i)

sol1 = sol[0]
sol2 = sol[1]
```

```python
sol1
```

```python
sol2
```

```python
sol1.subs(j,1).subs(k,1)
```

```python
sol2.subs(j,1).subs(k,1)
```
