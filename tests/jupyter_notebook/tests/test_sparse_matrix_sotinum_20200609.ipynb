```python
K = oti.lil_matso((100,100))
for i in range(K.shape[0]):
    K[i,i] = i+1+oti.e(i+1,order=2)
# end for 
K = K.tocsr()
```

```python
b = oti.ones((100,1))*3
```

```python
Kr_csr = K.real
```

```python
Kr_csr
```

```python
Kr_csc = Kr_csr.tocsc()
br = b.real
```

```python
%timeit c = spla.splu(Kr_csc).solve(br)
```
