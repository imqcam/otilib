# Test curvature and center of curvature

```python
import sys
path2oti = '../'
sys.path.append(path2oti) # Add path to OTI library.

# 2. OTI LIBRARY
import otilib as oti
np = oti.np
e = oti.e

# 6. PLOTTING
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
x=np.arange(0,2*np.pi,0.01)
```

```python
y=3.0*np.sin(x)



x_0 = np.pi*0.25 + e(1,order = 2)

y_0 = 3.0*oti.sin(x_0) 
```

```python
x_0r=x_0.getDeriv(0)
y_0r=y_0.getDeriv(0)


T = np.array([x_0.getDeriv([1]),y_0.getDeriv([1]) ])  # Get first and second order derivatives.
T = T/np.linalg.norm(T)

N = np.array([-T[1],T[0]])
N
```

```python
k = y_0.getDeriv([[1,2]])/(np.power(1.0+y_0.getDeriv([1])**2,3.0/2.0)) #curvature
k

rho = 1.0/k
c = np.array([x_0r,y_0r])+N*rho
```

```python
# compute circle:

xcirc = np.cos(x)*rho+c[0]
ycirc = np.sin(x)*rho+c[1]
```

```python
plt.figure()
plt.plot(x,y)
# Curve:
plt.plot([x_0r,x_0r+T[0]],[y_0r,y_0r+T[1]])

# Circle
plt.plot(c[0],c[1],'xg')
plt.plot(xcirc,ycirc,'g')
# Evaluation point
plt.plot([x_0r],[y_0r],'.r')
plt.axis('equal')
plt.grid('on')
plt.show()
```

```python
# Miranda Problem

T = np.array([-0.059892290728, 0.998204845466])


CT=np.array([-0.15,2.5])

du_dr = 1.1395E+08
d2u_dr2 =  2.8212E+09
du_ds = -6.5224E+06
d2u_ds2 = 2.1527E+09

k = d2u_dr2/(np.power(1.0+du_dr**2,3.0/2.0)) #curvature

N = np.array([-T[1],T[0]])

rho = 1.0/k
c = CT+N*rho

```
