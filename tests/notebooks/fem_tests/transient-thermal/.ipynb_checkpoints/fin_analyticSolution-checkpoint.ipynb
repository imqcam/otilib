# Parameters to insert in analytic solution (extracted from paper):

```python
# Data from paper.

b = 51.0/1000
δ = 4.75/1000

ρ    = 4430.0
Cp   =  580.0 
dt   =    1.0   
k    =    7.1   
hu   =  114.0 
Tinf =  283.0 
T0   =  283.0 
Tw   =  389.0 
Uv   =  735.0 

tmax = 350

neval = int(tmax/dt.real)
```

# Definition of analytic solution according to the paper:

```python
def analytic_solution(x, t, b, δ, 
                      ρ, Cp, dt, k, 
                      hu, T0, Tw, Tinf,
                      base = np,
                      nterms = 100 ):
    
    # Adimensionalization of input parameters
    θ0 = (T0 - Tinf)/(Tw - Tinf)
    X = x/b
    NN = ( 2 * hu * b**2 )/( k * δ )
    N = base.sqrt( NN )
    τ = ( t * k )/ ( b**2 * ρ * Cp )
    θτ  = 0
    
    θss = base.cosh( N * X )/base.cosh( N )
    
    for n in range(0,nterms+1):
        
        λn = ( ( 2. * n - 1.) / 2. ) * np.pi
        sign = ( -1 )**( n + 1 )
        
        t1 = θ0 / λn - ( λn**2 ) / ( NN + λn**2 )
        t2 = base.cos( λn * X )
        t3 = base.exp( -( NN + λn**2 ) * τ )
        
        θτ += sign * t1 * t2 * t3
        
    θτ *= 2
    
    θ = θss+θτ
    T = θ*(Tw-Tinf)+Tinf
    
    return T,θ,θτ,θss
```

# Evaluate analytic solution for different values of t.

```python
t = np.arange(dt,(neval+1)*dt,dt)
x = 0

Tanaly = np.zeros((t.size,))
θanaly = np.zeros((t.size,))

i=0
for ti in t:
    
    Ti,θi,θτi,θssi = analytic_solution(x, ti, b, δ, ρ, Cp, dt, k, hu, T0, Tw, Tinf,
                                   base = np,
                                   nterms = 1000 )
    Tanaly[i] = Ti
    θanaly[i] = θi
    i+=1
```

```python
fig,axs= plt.subplots(1,2,figsize=(9,5))

ax = axs[1]
ax.plot(t,θanaly,'-',label='\'Analytic\'')
ax.plot(tplot,θplot,label='FEM (same as paper)')

ax.axis([None,None,-0.001,0.04])
ax.set_xlabel("Time [s]")
ax.set_ylabel(r"$\theta$")
ax.set_title(r"$\theta$ focused in same range as paper")
ax.legend()
ax.grid()

ax = axs[0]

ax.plot(t,θanaly,'-',label='\'Analytic\'')
ax.plot(tplot,θplot,label='FEM (same as paper)')

ax.set_xlabel("Time [s]")
ax.set_ylabel(r"$\theta$")
ax.set_title(r"Behavior of $\theta$.")
ax.legend()
ax.grid()
plt.tight_layout()
plt.show()
```
