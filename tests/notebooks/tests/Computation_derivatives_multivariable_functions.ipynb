```python
str_funct_start = "// ****************************************************************************************************\n"
str_funct_end   = "// ----------------------------------------------------------------------------------------------------\n"
tab_sep = " "*2 # Four spaces.
endl="\n"

def write_bivar_procedure(string_out, fxy, x, y, deriv_order, level):
    
    # Initialize declarations:
    string_out += tab_sep*level
    string_out += "ord_t i;"+endl
        
    str_deriv_comp = ""
    str_deriv_comp += tab_sep*level
    str_deriv_comp += "for(i=0;i<=order;i++){"+endl
    level += 1

    str_deriv_comp += tab_sep*level
    str_deriv_comp += "switch(i){"+endl
    level += 1    
    
    
    declarations = []

    computed = {}
    for key in computations:
        computed[key] = False
    # end 
    k=0

    for ordi in range(0,deriv_order+1):
        
        print(" -- Writing derivative ",ordi)
        str_deriv_comp += tab_sep*level
        str_deriv_comp += "case {0}:\n".format(ordi)
        level += 1
        
        for dy in range(ordi+1):
            
            dx = ordi-dy
            str_deriv_comp += tab_sep*level
            str_deriv_comp += '//d^{0}f/dx^{1}dy^{2}:\n'.format(ordi,dx,dy)
            
            deriv = fxy.diff(x,dx).diff(y,dy).evalf()
            ccode = sym.ccode(deriv).replace(" ","")+";"
            
            # Perform known substitutions:
            for key,value in substitutions.items():

                if key in ccode:
                    
                    ccode = ccode.replace(key,value)
                    
                    if not computed[value] :
                        str_deriv_comp += tab_sep*level
                        str_deriv_comp += value+"="+computations[value]+";\n"
                        computed[value] = True
                        declarations.append(value)
                    # end if 
                    
                # end if 
                
            # end for
            
            # Clean ccode.
            ccode = ccode.replace(".0+",".+")
            ccode = ccode.replace(".0-",".-")
            ccode = ccode.replace(".0*",".*")
            ccode = ccode.replace(".0/","./")
            ccode = ccode.replace(".0,",".,")
            ccode = ccode.replace(".0;",".;")
            ccode = ccode.replace(".0)",".)")
            if ccode[-1]!=";":
                ccode += ";"
            # end if
            str_deriv_comp += tab_sep*level
            str_deriv_comp += 'derivs[{0}]={1}\n'.format(k,ccode)
            k+=1
        level -= 1

    # end for 
    


    # ending SWITCH.
    level-=1
    str_deriv_comp += tab_sep*level
    str_deriv_comp += "}"+endl


    # ending FOR.
    level-=1
    str_deriv_comp += tab_sep*level
    str_deriv_comp += "}"+endl
    
    
    # Append declarations:
    k = 1
    string_out += tab_sep*level
    string_out += "coeff_t "
    for decl in declarations:
        
        if k%10 == 0:
            string_out  = string_out[:-1]
            string_out += ";\n"
            string_out += tab_sep*level
            string_out += "coeff_t "
        # end if 
        
        string_out+=decl + ','
        
        
        
        k += 1
    
    # end for 
    string_out  = string_out[:-1]
    string_out += ";\n"
    
    string_out += str_deriv_comp
    
    return string_out

def write_singvar_procedure(string_out, fx, x, deriv_order, level):
    
    # Initialize declarations:
    string_out += tab_sep*level
    string_out += "ord_t i;"+endl
        
    str_deriv_comp = ""
    str_deriv_comp += tab_sep*level
    str_deriv_comp += "for(i=0;i<=order;i++){"+endl
    level += 1

    str_deriv_comp += tab_sep*level
    str_deriv_comp += "switch(i){"+endl
    level += 1    
    
    
    declarations = []

    
    computed = {}
    for key in computations:
        computed[key] = False
    # end 
    k=0

    for ordi in range(0,deriv_order+1):
        
        print(" -- Writing derivative ",ordi)
        str_deriv_comp += tab_sep*level
        str_deriv_comp += "case {0}:\n".format(ordi)
        level += 1
        
        str_deriv_comp += tab_sep*level
        str_deriv_comp += '//d^{0}f/dx^{0}:\n'.format(ordi)
        

        deriv = fx.diff(x,ordi).evalf()
        ccode = sym.ccode(deriv).replace(" ","")+";"

        # Perform known substitutions:
        for key,value in substitutions.items():

            if key in ccode:

                ccode = ccode.replace(key,value)

                if not computed[value] :
                    str_deriv_comp += tab_sep*level
                    str_deriv_comp += value+"="+computations[value]+";\n"
                    computed[value] = True
                    declarations.append(value)
                # end if 

            # end if 

        # end for

        # Clean ccode.
        ccode = ccode.replace(".0+",".+")
        ccode = ccode.replace(".0-",".-")
        ccode = ccode.replace(".0*",".*")
        ccode = ccode.replace(".0/","./")
        ccode = ccode.replace(".0,",".,")
        ccode = ccode.replace(".0;",".;")
        ccode = ccode.replace(".0)",".)")
        if ccode[-1]!=";":
            ccode += ";"
        # end if
        str_deriv_comp += tab_sep*level
        str_deriv_comp += 'derivs[{0}]={1}\n'.format(k,ccode)
        k+=1
        level -= 1

    # end for 
    


    # ending SWITCH.
    level-=1
    str_deriv_comp += tab_sep*level
    str_deriv_comp += "}"+endl


    # ending FOR.
    level-=1
    str_deriv_comp += tab_sep*level
    str_deriv_comp += "}"+endl
    
    
    # Append declarations:
    k = 1
    string_out += tab_sep*level
    string_out += "coeff_t "
    for decl in declarations:
        
        if k%10 == 0:
            string_out  = string_out[:-1]
            string_out += ";\n"
            string_out += tab_sep*level
            string_out += "coeff_t "
        # end if 
        
        string_out+=decl + ','
        
        
        
        k += 1
    
    # end for 
    string_out  = string_out[:-1]
    string_out += ";\n"
    
    string_out += str_deriv_comp
    
    return string_out
```
