```python
import pyoti.sparse as oti
import numpy as np
```

```python
x = np.pi/6 + oti.e(1, order=50)
y = 2*np.pi/6 + oti.e(2, order=50)

f = oti.sin(oti.exp((x+y)**2))
f
```

```python
f.get_deriv([[1,50]])
```

```python
import sympy as sym
```

```python
xs = sym.symbols('x')
fs = sym.sin(sym.exp(xs**2))
dfs = fs.diff(xs,50)
```
