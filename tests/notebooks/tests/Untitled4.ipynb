```python

t = np.arange(1,1000,1)
t2 = t*10
y1 = t**1
y2 = t**2
y3 = t**3
y4 = (10*t)**3
```

```python
plt.figure(figsize=(5,8))

plt.loglog(t,y1,label='t**1')
plt.loglog(t,y2,label='t**2')
plt.loglog(t,y3,label='t**3')
plt.loglog(t2,y3,label='other')

plt.legend()
plt.gca().minorticks_on()
plt.gca().set_axisbelow(True)
plt.grid()
plt.grid(b=True,which='minor')
plt.gca().yaxis.set_major_locator(LogLocator(10,numticks=30))


plt.plot()
```
