```python

ctypes ={}
ctypes['F']='fearrso'
ctypes['O']='arrso'
ctypes['R']='darr'
ctypes['f']='fesoti'
ctypes['o']='sotinum'
ctypes['r']='coeff'

cfunc_prefix ={}
cfunc_prefix['F']='fearrso'
cfunc_prefix['O']='arrso'
cfunc_prefix['R']='darr'
cfunc_prefix['f']='fesoti'
cfunc_prefix['o']='soti'
cfunc_prefix['r']=''

cvars ={}
cvars['F']='cF'
cvars['O']='cO'
cvars['R']='cR'
cvars['f']='cf'
cvars['o']='co'
cvars['r']='cr'


pytypes ={}
pytypes['F']='matsofe'
pytypes['O']='matso'
pytypes['R']='dmat'
pytypes['f']='sotife'
pytypes['o']='sotinum'
pytypes['r']='coeff_t'



pyop ={}
pyop['sum']='add'
pyop['sub']='sub'
pyop['mul']='mul'
pyop['div']='div'

opname ={}
opname['sum']='Addition'
opname['sub']='Subtraction'
opname['mul']='Multiplication'
opname['div']='Division'

opsymbol ={}
opsymbol['sum']='+'
opsymbol['sub']='-'
opsymbol['mul']='*'
opsymbol['div']='/'

pyvars ={}
pyvars['F']='F'
pyvars['O']='O'
pyvars['R']='R'
pyvars['f']='f'
pyvars['o']='o'
pyvars['r']='r'

def t_res(t_lhs, t_rhs):
    if t_lhs == 'F' :
        return 'F'
    elif t_rhs=='F':
        return 'F'
    elif t_lhs == 'f' and (t_rhs=='O' or t_rhs=='R'):
        return 'F'
    elif t_rhs == 'f' and (t_lhs=='O' or t_lhs=='R'):
        return 'F'
    
    elif t_lhs == 'f' or t_rhs == 'f':
        return 'f'
    
    elif t_rhs=='O' or t_lhs=='O':
        return 'O'
    
    elif t_lhs=='o' and t_rhs=='R':
        return 'O'
    
    elif t_rhs=='o' and t_lhs=='R':
        return 'O'
    
    elif t_rhs=='o' or t_lhs=='o':
        return 'o'
    
    elif t_rhs=='R' or t_lhs=='R':
        return 'R'
    
    else:
        return 'r'
    # end if 
# end function

def arr_num(typ):
    if typ == 'F' or typ == 'O' or typ == 'R' :
        return 'arr'
    else:
        return 'num'
    # end if 
# end function

def flip_lhs_rhs(lhs,op,rhs):
    if op == 'sum' or op == 'mul':
        if arr_num(lhs)=='arr' and arr_num(rhs)=='num':
            return True
        elif lhs=='F' and (rhs=='O' or rhs=='R'):
            return True
        elif lhs=='O' and (rhs=='R'):
            return True
        elif lhs=='f' and (rhs=='o' or rhs=='r'):
            return True
        elif lhs=='o' and (rhs=='r'):
            return True
        # end if
    # end if 
    return False
# end function


def is_supported(lhs,op,rhs):
    if op == 'sum' or op == 'mul' or op == 'sub' or op == 'div':
        res = t_res(lhs, rhs)
        if res != lhs and res != rhs:
            return False
        # end if 
    # end if 
    return True
# end function

def get_pydata(typ,post=''):
    if typ != 'r':
        return "&"+pvars[typ]+post+"."+arr_num(typ)
    else:
        return pvars[typ]+post
    # end if 
# end function
```

```python
def write_funct_to(lhs,rhs,indent = "", op = 'sum'):
    
    res = t_res(lhs,rhs)
    
    out = ''
    resvar = pyvars[res] + "res"
    out += indent + resvar + ' = out\n'
    out += indent 
    
    if flip_lhs_rhs(lhs,op,rhs):
        out += cfunc_prefix[res]+'_'+op+'_'+rhs+lhs+"_to( "
        out += get_pydata(rhs,post='rhs')+', ' 
        out += get_pydata(lhs,post='lhs')+', ' 
        out += get_pydata(res,post='res')
    else:
        out += cfunc_prefix[res]+'_'+op+'_'+lhs+rhs+"_to( "
        out += get_pydata(lhs,post='lhs') +', ' 
        out += get_pydata(rhs,post='rhs') +', '
        out += get_pydata(res,post='res')
    # end if 
    
    if res != 'R' and res != 'r':
        out += ', dhl'
    # end if 
    
    out += " )\n"
        
    return out
# end function

def write_funct(lhs,rhs,indent = "", op = 'sum'):
    res = t_res(lhs,rhs)
    out = ''
    resvar = cvars[res] + "res"
    out += indent + resvar + ' = '
    
    if flip_lhs_rhs(lhs,op,rhs):
        out += cfunc_prefix[res]+'_'+op+'_'+rhs+lhs+"( "
        out += get_pydata(rhs,post='rhs')+', ' 
        out += get_pydata(lhs,post='lhs') 
    else:        
        out += cfunc_prefix[res]+'_'+op+'_'+lhs+rhs+"( "
        out += get_pydata(lhs,post='lhs') +', ' 
        out += get_pydata(rhs,post='rhs')
    # end if 

    if res != 'R' and res != 'r':
        out += ', dhl'
    # end if 
    
    out += " )\n"
    out += indent+'res = ' + pytypes[res]+'.create( &' + resvar + " )\n"
        
    return out
# end function
```

```python
def write_cython_algebra_helper_lhs(lhs,op = "sum", T_rhs = ["o","f",'F','O','R','r'],level=0, spaces=2):
        
    indent = (level*spaces)*' '

    out = indent+'#'+'*'*(102-len(indent)-1)+'\n'
    out += indent+'cdef object __'+pyop[op]+'__'+pyvars[lhs]+'X__('+pytypes[lhs]+' '+pyvars[lhs]+'lhs, object rhs, object out = None):\n'

    level+=1; indent = (level*spaces)*' '


    out += indent + '"""\n'
    out += indent + 'PORPUSE: '+opname[op]+' between '+pytypes[lhs]+' and object.\n\n'
    out += indent + '"""\n'

    out += indent+'#'+'*'*(102-len(indent)-1)+'\n\n'

    out += indent+'global dhl\n\n'

    out += indent+'# Scalar types:\n'
    out += indent + "cdef sotinum    orhs, ores\n"
    out += indent + "cdef sotinum_t cores\n\n"
    out += indent + "cdef sotife     frhs, fres\n"
    out += indent + "cdef fesoti_t  cfres\n\n"
    out += indent + "cdef coeff_t    rrhs, rres\n"
    out += indent + "cdef coeff_t   crres\n\n"

    out += indent+"# Array types\n"
    out += indent+"cdef matso      Orhs, Ores\n"
    out += indent+"cdef arrso_t   cOres\n\n"
    out += indent+"cdef dmat       Rrhs, Rres\n"
    out += indent+"cdef darr_t    cRres\n\n"
    out += indent+"cdef matsofe    Frhs, Fres\n"
    out += indent+"cdef fearrso_t cFres\n\n"

    out += indent+"cdef uint8_t res_flag = 1\n"
    out += indent+"cdef object  res = None\n\n"

    out += indent+"trhs = type(rhs)\n\n"

    out += indent+"if out is None:\n"

    level+=1; indent = (level*spaces)*' '

    out += indent+"res_flag = 0\n"

    level-=1; indent = (level*spaces)*' '  

    out += indent+"# end if\n\n"

    i=0
    
    for rhs in T_rhs:

        if is_supported(lhs,op,rhs):
            if i == 0:
                start = "if"
                i+=1
            else:
                start="elif"
            # end 

            if rhs == 'r':
                out += indent + start + ' trhs in number_types:\n'
            else:
                out += indent + start + ' trhs == ' + pytypes[rhs] + ':\n'
            # end if 

            level+=1; indent = (level*spaces)*' '
            
            if lhs == 'r' and rhs == 'r':
                out += indent + pyvars[rhs]+'rhs = rhs\n'
                out += indent + 'res = '+pyvars[lhs]+'lhs'+opsymbol[op]+pyvars[rhs]+'rhs\n'
                level -= 1; indent = (level*spaces)*' '
                
            else:
                out += indent + pyvars[rhs]+'rhs = rhs\n'
                out += indent + "if res_flag:\n"

                level+=1; indent = (level*spaces)*' '

                out += write_funct_to(lhs,rhs,indent=indent, op=op)    

                level-=1; indent = (level*spaces)*' '

                out += indent + "else:\n"

                level+=1; indent = (level*spaces)*' '

                out += write_funct(lhs,rhs,indent=indent, op=op)

                level-=1; indent = (level*spaces)*' '

                out += indent + "# end if\n"

                level -= 1; indent = (level*spaces)*' '
                
            # end if 

        # end if 
        
    # end for
    indent = (level*spaces)*' ' 

    out += indent+"else:\n"

    level+=1; indent = (level*spaces)*' '

    out += indent+'raise TypeError("Unsupported '+pyop[op]+' operation between {0} and {1}.".format(type('+pyvars[lhs]+'lhs),trhs))\n'

    level-=1; indent = (level*spaces)*' '

    out += indent+'# end if\n\n'

    out += indent+"if res_flag == 0:\n"
    level+=1; indent = (level*spaces)*' '
    out += indent+"return res\n"    
    level-=1; indent = (level*spaces)*' '
    out += indent+"# end if\n\n"

    level-=1; indent = (level*spaces)*' '  

    out += indent+'#'+'-'*(102-len(indent)-1)+'\n'
    
    return out

# end function

```
